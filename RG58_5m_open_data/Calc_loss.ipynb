```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fname = "open_rl_data.s1p"
freq = np.loadtxt(fname, usecols=0)*1e-6
S11_cable = np.loadtxt(fname, usecols=1) + np.loadtxt(fname, usecols=2)*1j
```

```python
plt.figure(figsize=(16,9))
plt.plot(freq, 20*np.log10(np.abs(S11_cable)))

plt.figure(figsize=(16,9))
plt.plot(freq, np.unwrap(np.angle(S11_cable)))
```

```python
wl = 300/freq
l = 5/(0.66)
alpha = 0.03/wl
beta = 2*np.pi/wl
gamma = alpha+beta*1j

Z_in = 50*(1+np.exp(-2*gamma*l))/(1-np.exp(-2*gamma*l))
gamma_sim = (Z_in-50)/(Z_in+50)

# plt.figure(figsize=(16,9))
# plt.plot(freq, np.real(S11_cable))
# plt.plot(freq, np.real(gamma_sim))

# plt.figure(figsize=(16,9))
# plt.plot(freq, np.imag(S11_cable))
# plt.plot(freq, np.imag(gamma_sim))

# plt.figure(figsize=(16,9))
# plt.plot(freq, np.abs(S11_cable))
# plt.plot(freq, np.abs(gamma_sim))

# plt.figure(figsize=(16,9))
# plt.plot(freq, np.imag(S11_cable))
# plt.plot(freq, np.imag(gamma_sim))

plt.figure(figsize=(16,9))
plt.plot(freq, 20*np.log10(np.abs(np.fft.ifft(S11_cable))))
plt.plot(freq, 20*np.log10(np.abs(np.fft.ifft(gamma_sim))))
plt.xlim([0,25])
plt.stem([5.5],[0])
plt.stem([11],[0])
```
