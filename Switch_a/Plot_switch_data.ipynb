```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fname = "Com2NC.s2p"
freq = np.loadtxt(fname, usecols=0)*1e-6
S11_Com2NC = np.loadtxt(fname, usecols=1) + np.loadtxt(fname, usecols=2)*1j
S21_Com2NC = np.loadtxt(fname, usecols=3) + np.loadtxt(fname, usecols=4)*1j

fname = "Com2NO_ON.s2p"
freq = np.loadtxt(fname, usecols=0)*1e-6
S11_Com2NO_ON = np.loadtxt(fname, usecols=1) + np.loadtxt(fname, usecols=2)*1j
S21_Com2NO_ON = np.loadtxt(fname, usecols=3) + np.loadtxt(fname, usecols=4)*1j

plt.figure(figsize=(16,9))
plt.plot(freq, 20*np.log10(np.abs(S21_Com2NC)))
plt.plot(freq, 20*np.log10(np.abs(S21_Com2NO_ON)))

plt.figure(figsize=(16,9))
plt.plot(freq, np.angle(S21_Com2NC))
plt.plot(freq, np.angle(S21_Com2NO_ON))
```
