```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH = "/Users/jishnu/codespace/AMATEUR_RADIO/Code_nanovna/RG58_cable/Stub/"
file_name_stub_open = "cable_stub_test_2_open.s1p"
file_name_stub_short = "cable_stub_test_2_short.s1p"
```

```python
freq = np.loadtxt(PATH+file_name_stub_open, usecols=0)/1e6

S11_c_stub_open = np.loadtxt(PATH+file_name_stub_open, usecols=1) + np.loadtxt(PATH+file_name_stub_open, usecols=2)*1j
S11_c_stub_short = np.loadtxt(PATH+file_name_stub_short, usecols=1) + np.loadtxt(PATH+file_name_stub_short, usecols=2)*1j

S21_c_stub_open = np.loadtxt(PATH+file_name_stub_open, usecols=3) + np.loadtxt(PATH+file_name_stub_open, usecols=4)*1j
S21_c_stub_short = np.loadtxt(PATH+file_name_stub_short, usecols=3) + np.loadtxt(PATH+file_name_stub_short, usecols=4)*1j
```

```python
plt.figure(figsize=(16,9))
plt.plot(freq, 20*np.log10(np.abs(S11_c_stub_open)), label="Open stub")
plt.plot(freq, 20*np.log10(np.abs(S11_c_stub_short)), label="Shorted stub")
plt.legend(loc='best')
plt.xlim([min(freq), max(freq)])
plt.grid()
plt.xlabel("Freq (MHz)")
plt.ylabel("S11 dB")

plt.figure(figsize=(16,9))
plt.plot(freq, 20*np.log10(np.abs(S21_c_stub_open)), label="Open stub")
plt.plot(freq, 20*np.log10(np.abs(S21_c_stub_short)), label="Shorted stub")
plt.legend(loc='best')
plt.xlim([min(freq), max(freq)])
plt.grid()
plt.xlabel("Freq (MHz)")
plt.ylabel("S21 dB")
```
