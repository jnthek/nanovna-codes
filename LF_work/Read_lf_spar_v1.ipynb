```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
S11_file = "ferrite_loop_lowfreq_2.s1p"

freq = np.loadtxt(S11_file, usecols=0)/1e3
S11 = np.loadtxt(S11_file, usecols=1)+np.loadtxt(S11_file, usecols=2)*1j

imped = 50*(1+S11)/(1-S11)
```

```python
L = 1e-3
C = 100*1e-12
R = 1000
XC = 1.0/(2j*np.pi*freq*1e3*C)
XL = (2j*np.pi*freq*1e3*L)

coil_imp = XL+R

X = coil_imp*XC/(coil_imp+XC)

plt.figure()
plt.plot(freq, np.abs(S11))
plt.plot(freq, np.real(imped))
plt.plot(freq, np.imag(imped))

plt.plot(freq, np.imag(X))
```
