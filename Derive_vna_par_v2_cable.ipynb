```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
```

```python
PATH = "./CABLE_DATA/"
freq = np.loadtxt(PATH+"lc_short_S11.s1p", usecols=0)/1e3
Gamma_open = np.loadtxt(PATH+"lc_open_S11.s1p", usecols=1)+np.loadtxt(PATH+"lc_open_S11.s1p", usecols=2)*1j
Gamma_short = np.loadtxt(PATH+"lc_short_S11.s1p", usecols=1)+np.loadtxt(PATH+"lc_short_S11.s1p", usecols=2)*1j
Gamma_load = np.loadtxt(PATH+"lc_load_S11.s1p", usecols=1)+np.loadtxt(PATH+"lc_load_S11.s1p", usecols=2)*1j
```

```python
plt.figure(figsize=(16,9))
plt.plot(freq, 20*np.log10(np.abs(Gamma_open)), label="Open")
plt.plot(freq, 20*np.log10(np.abs(Gamma_short)), label="Short")
plt.plot(freq, 20*np.log10(np.abs(Gamma_load)), label="Load")
plt.legend()
plt.xlim([min(freq), max(freq)])
plt.grid()
```

```python
ed = Gamma_load
ep = (Gamma_open+Gamma_short-2*Gamma_load)/(Gamma_open-Gamma_short)
et_a = (Gamma_open-Gamma_short-ep*(Gamma_open+Gamma_short))/2
et = et_a+(ep*ed)
```

```python
plt.figure(figsize=(16,9))
plt.plot(freq, 20*np.log10(np.abs(et)), label="et")
plt.plot(freq, 20*np.log10(np.abs(ep)), label="ep")
plt.plot(freq, 20*np.log10(np.abs(ed)), label="ed")
plt.legend()
plt.xlim([min(freq), max(freq)])
plt.grid()
```
